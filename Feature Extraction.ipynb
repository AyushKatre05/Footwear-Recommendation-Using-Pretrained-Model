```python
import cv2
import numpy as np
from numpy.linalg import norm

def extract_feature(img_path, model):
    img = cv2.imread(img_path)
    img = cv2.resize(img, (720, 540))
    img = np.array(img)
    expand_img = np.expand_dims(img, axis=0)
    pre_img = preprocess_input(expand_img)
    result = model.predict(pre_img).flatten()  # Corrected the flatten call
    normalized = result / norm(result)
    return normalized

# Call the function
extract_feature("Footwear/Men/Images/images_with_product_ids/1636.jpg", model)
```

```python
import os
import cv2
import numpy as np
from tqdm import tqdm

# Define the extract_feature function
def extract_feature(img_path, model):
    try:
        img = cv2.imread(img_path)
        if img is None:
            print(f"Error: Unable to read image '{img_path}'. Skipping...")
            return None
        img = cv2.resize(img, (224, 224))  # Resize to match ResNet50 input size
        img = np.array(img)
        expand_img = np.expand_dims(img, axis=0)
        pre_img = preprocess_input(expand_img)
        result = model.predict(pre_img).flatten()
        normalized = result / np.linalg.norm(result)  # Use np.linalg.norm for calculating norm
        return normalized
    except Exception as e:
        print(f"Error processing '{img_path}': {e}")
        return None

# Define data directories for different categories
men_directory = r'C:\Users\ayush\OneDrive\Desktop\Footwear Recommendation System\Footwear\Men\Images\images_with_product_ids'
women_directory = r'C:\Users\ayush\OneDrive\Desktop\Footwear Recommendation System\Footwear\Women\Images\images_with_product_ids'

# List of directories containing images
data_directories = [men_directory, women_directory]

men_filenames = []
women_filenames = []

# Collect paths of all images in the specified directories for each category
for directory in data_directories:
    for root, _, files in os.walk(directory):
        for file in files:
            full_path = os.path.join(root, file)
            if directory == men_directory:
                men_filenames.append(full_path)
            elif directory == women_directory:
                women_filenames.append(full_path)

# Display the first 5 filenames from the men_filenames list
print("Men's Filenames:")
for file in men_filenames[:5]:
    print(file)

# Display the first 5 filenames from the women_filenames list
print("\nWomen's Filenames:")
for file in women_filenames[:5]:
    print(file)

# Initialize feature_list
feature_list = []

# Iterate over all paths and extract features
for filename in tqdm(men_filenames + women_filenames):
    feature = extract_feature(filename, model)
    if feature is not None:
        feature_list.append(feature)

# Save feature_list and filenames to pickle files
import pickle
pickle.dump(feature_list, open('featurevector.pkl', 'wb'))
pickle.dump(men_filenames + women_filenames, open('filename.pkl', 'wb'))
```

```python
import os
import pickle

# Define data directories for different categories
men_directory = r'C:\Users\ayush\OneDrive\Desktop\Footwear Recommendation System\Footwear\Men\Images\images_with_product_ids'
women_directory = r'C:\Users\ayush\OneDrive\Desktop\Footwear Recommendation System\Footwear\Women\Images\images_with_product_ids'

# List of directories containing images
data_directories = {'men': men_directory, 'women': women_directory}

# Collect paths of all images in the specified directories for each category
filenames = {}
for category, directory in data_directories.items():
    filenames[category] = []
    for root, _, files in os.walk(directory):
        for file in files:
            full_path = os.path.join(root, file)
            filenames[category].append(full_path)

# Save filenames to pickle files
pickle.dump(filenames['men'], open('men_filenames.pkl', 'wb'))
pickle.dump(filenames['women'], open('women_filenames.pkl', 'wb'))
```

```python
data_directories = {'men': men_directory, 'women': women_directory}

# Collect feature vectors for each category
feature_vectors = {}
for category, directory in data_directories.items():
    feature_vectors[category] = []
    for root, _, files in os.walk(directory):
        for file in tqdm(files):
            full_path = os.path.join(root, file)
            feature = extract_feature(full_path, model)
            if feature is not None:
                feature_vectors[category].append(feature)

# Save feature vectors to pickle files
pickle.dump(feature_vectors['men'], open('men_features.pkl', 'wb'))
pickle.dump(feature_vectors['women'], open('women_features.pkl', 'wb'))
```
